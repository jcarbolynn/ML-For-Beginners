```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

# load diabetes dataset, X is data, y is target
X, y = datasets.load_diabetes(return_X_y=True)

print(y)

# print shape of data and first row
# print(X.shape)
# print(X[0])
# X[:3]
# print(type(X))
# print(X.dtype.names)
```

```python
# extract column at index 2 (bmi)
X = X[:, 2]
print(X.shape)

# reshape into 2D array
X = X.reshape((-1,1))
print(X.shape)
```

```python
# split into training and testing data
# 2/3 for training, 1/3 for testing
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# create linear regression model and train with our data
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# now make predition using test data
y_pred = model.predict(X_test)
```

```python
# now check to see how well our predictions were

# create scatterplot
plt.scatter(X_test, y_test, color='black')

# plot predictions
plt.plot(X_test, y_pred, color = 'blue', linewidth = 3)

# add labels and title
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('Graph showing Diabetes Progression vs BMI')

# draw the plot
plt.show()
```
